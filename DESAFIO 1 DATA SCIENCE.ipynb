### Importación de datos



```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1= pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



1.Ingreso total por cada tienda

```python
ingreso_tienda1 = tienda1["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()
```

```python
print("Ingreso total de cada tienda:")
print(f"Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda4:,.2f}")
```

**GRAFICO**

```python
# Ingresos por tienda
ingresos = [
    tienda1["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tiendas")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

# 2. Ventas por categoría

```python
# Tienda 1 - contar productos por categoría
print("Productos vendidos por categoría - Tienda 1:")
print(tienda1["Categoría del Producto"].value_counts())
print()
```

**GRAFICO**

```python
# Contar los productos vendidos por categoría
categorias_t1 = tienda1["Categoría del Producto"].value_counts()

# Total de artículos vendidos
total_articulos = categorias_t1.sum()

# Función personalizada para mostrar cantidades reales en lugar de porcentajes
def mostrar_cantidad(porcentaje):
    cantidad = int(round(porcentaje * total_articulos / 100))
    return f'{cantidad} artículos'

# Crear gráfico de pastel con cantidades
plt.figure(figsize=(7,7))
plt.pie(
    categorias_t1,
    labels=categorias_t1.index,
    autopct=mostrar_cantidad,
    startangle=140
)
plt.title(f"Distribución de Categorías - Tienda 1\nTotal de artículos vendidos: {total_articulos}")
plt.axis("equal")
plt.show()
```

```python
# Tienda 2 - contar productos por categoría
print("Productos vendidos por categoría - Tienda 2:")
print(tienda2["Categoría del Producto"].value_counts())
print()
```

**GRAFICO**

```python
categorias_t2 = tienda1["Categoría del Producto"].value_counts()
plt.figure(figsize=(7,7))
plt.pie(categorias_t2, labels=categorias_t1.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribución de Categorías - Tienda 2")
plt.axis("equal")  # Círculo perfecto
plt.show()
```

```python
# Tienda 3 - contar productos por categoría
print("Productos vendidos por categoría - Tienda 3:")
print(tienda3["Categoría del Producto"].value_counts())
print()
```

```python
# Tienda 4 - contar productos por categoría
print("Productos vendidos por categoría - Tienda 4:")
print(tienda4["Categoría del Producto"].value_counts())
print()
```

# 3. Calificación promedio de la tienda


```python
promedio1 = tienda1["Calificación"].mean()
promedio2 = tienda2["Calificación"].mean()
promedio3 = tienda3["Calificación"].mean()
promedio4 = tienda4["Calificación"].mean()
```

```python
print("Calificación promedio de los clientes por tienda:")
print(f"Tienda 1: {promedio1:.2f}")
print(f"Tienda 2: {promedio2:.2f}")
print(f"Tienda 3: {promedio3:.2f}")
print(f"Tienda 4: {promedio4:.2f}")
```

**GRAFICO**

```python
# Promedios de calificación
calificaciones = [
    tienda1["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

plt.figure(figsize=(8,5))
plt.plot(tiendas, calificaciones, marker='o', linestyle='-', color='green')
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación (1 a 5)")
plt.xlabel("Tiendas")
plt.ylim(0, 5.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# **Informe Final – Recomendación para el Sr. Juan**

**Introducción**
El objetivo de este análisis es identificar cuál de las cuatro tiendas disponibles representa la mejor opción para que el Sr. Juan venda sus productos. Para tomar esta decisión, se analizaron datos reales de ventas en cada tienda, considerando los siguientes factores clave:

- Ingresos totales generados.

- Categorías de productos más y menos vendidas.

- Calificaciones promedio de los clientes.

- Productos más y menos vendidos.

- Costo de envío promedio.

A partir de estos datos, se generaron gráficos que nos ayudan a visualizar y comparar el desempeño de cada tienda. A continuación, se presentan los hallazgos más relevantes.

**Análisis de Resultados**


*   Ingresos Totales:
   
    El gráfico de barras mostró que la Tienda 3 fue la que generó los mayores ingresos totales, seguida por la Tienda 2. Las tiendas 1 y 4 tuvieron ingresos más bajos. Esto indica una alta demanda en la Tienda 3.

*  Categorías más vendidas

   En el gráfico de pastel de la Tienda 1, vimos una alta concentración en categorías específicas, como Tecnología o Moda, lo cual puede ser una oportunidad si los productos del Sr. Juan se alinean con esas categorías. También se mostró el número exacto de artículos vendidos por categoría, lo cual facilita entender el volumen de ventas.


*   Calificaciones de los Clientes:

    Al comparar las calificaciones promedio, la Tienda 2 tuvo la mejor puntuación (más cerca de 5), lo cual refleja una alta satisfacción del cliente. Aunque la Tienda 3 genera más ingresos, su calificación fue ligeramente inferior.

*   Productos más y menos vendidos:

    Se identificaron los productos más populares y los menos vendidos por tienda. Esto ayuda a saber si hay productos con poca rotación que podrían afectar las ventas, o si hay productos exitosos que pueden complementar lo que vende el Sr. Juan.

   
*   Costo de Envío Promedio:

    El análisis mostró que la Tienda 1 tiene el costo de envío más bajo, mientras que la Tienda 4 tiene un costo mayor. Un menor costo de envío puede ser atractivo tanto para el vendedor como para los clientes, aunque no debe ser el único criterio para decidir.

**Conclusión y Recomendación Final**

Después de analizar todos los factores, recomendamos que el Sr. Juan venda en la Tienda 3 por las siguientes razones:

 - Es la tienda con los mayores ingresos totales, lo que indica un alto volumen de ventas.

- Tiene una variedad de productos exitosos, lo que sugiere un mercado activo.

- Aunque su calificación es ligeramente más baja que la de la Tienda 2, sigue siendo aceptable, y se puede mejorar con buen servicio y productos de calidad.

- El costo de envío no es el más bajo, pero está dentro de un rango razonable.

-> Tienda 2 sería una segunda opción viable si el Sr. Juan prioriza la reputación y la experiencia del cliente, aunque sus ingresos son algo menores.
